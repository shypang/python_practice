```python
import pandas as pd
```

```python
df1= pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},
    index=[0,1,2,3])
df2= pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},
    index=[4,5,6,7])
df3= pd.DataFrame({
    'A':['A8','A9','A10','A11'],
    'B':['B8','B9','B10','B11'],
    'C':['C8','C9','C10','C11'],
    'D':['D8','D9','D10','D11']},
    index=[8,9,10,11])
```

```python
result=pd.concat([df1,df2,df3])
result
```

```python
result=pd.concat([df1,df2,df3],keys=['x','y','z']) #인덱스값주고 병합 
result
```

```python



df4= pd.DataFrame({                     #같은것끼리 병합되고 같은 값이 없을경우 없는값
    'B':['B2','B3','B6','B7'], 
    'D':['D2','D3','D6','D7'],
    'F':['F2','F3','F6','F7']},
    index=[2,3,6,7])
result=pd.concat([df1,df4],axis=1)

```

```python
result
```

```python


result=pd.concat([df1,df4],axis=1,join='inner')# 같은것끼리 조인하여 나타냄 
result

```

```python


result=pd.concat([df1,df4],axis=1,join='inner')# 같은것끼리 조인하여 나타냄 
result

```

```python
result=pd.concat([df1,df4],axis=1,join_axes=[df1.index])# df1 값을 중심으로 전부나타나줄수 있게해줌 

result
```

```python
result=pd.concat([df1,df4],sort=False) #axis=0 , axis값 정의 안되있으면 행값이 기본값 
                                       #행값으로 그냥 병합 시키기

result
```
